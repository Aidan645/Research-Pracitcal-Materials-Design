```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
## Change this to ~1500 when saving for the last time (saves time while running to debug but change to 1500 to get high res images)
dpi_for_saving = 900

## if you get errors because have no latex in your computer you can make this False
plt.rcParams['text.usetex'] = True
```

```python
df = pd.read_csv("Feret Distribution Grain.csv")
fig,ax = plt.subplots()
bin_starts = df["bin start"].values
counts = df["count"].values
# Calculate bin ends
bin_ends = bin_starts + bin_starts[1]-bin_starts[0]

# Plot histogram
ax.bar(bin_starts, counts, width=[end - start for start, end in zip(bin_starts, bin_ends)], align='edge')
ax.set(title = "Histrogram of grain sizes for BTO sample of group 1",ylabel = "Counts", xlabel = "Feret Diameter [um]")
fig.savefig(dpi = dpi_for_saving,fname = "Domain Histo",bbox_inches='tight')
fig.show()
```
